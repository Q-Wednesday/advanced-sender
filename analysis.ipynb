```python
recv_timestamps = {}
send_timestamps= {}
for speed in [50,100,200]:
    recv_timestamps[speed]=get_timestamps("recv_{}.pcapng".format(speed))
    send_timestamps[speed]=get_timestamps("send_{}.pcapng".format(speed))
    print("Speed {}, send: {} packets, receive {} packets".format(speed,len(send_timestamps[speed]),len(recv_timestamps[speed])))
    print("Max interval: send:{}, receive:{}".format(max_gap(send_timestamps[speed]),max_gap(recv_timestamps[speed])))
```

```python
plt.figure(figsize=(16,4))
scatter_2data(send_timestamps[50],recv_timestamps[50])
```

```python
plt.figure(figsize=(16,4))
scatter_2data(specified_period(send_timestamps[50],0.050),specified_period(recv_timestamps[50],0.050))
```
